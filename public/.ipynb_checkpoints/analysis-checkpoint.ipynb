# Useful methods

```python
# Write column names

def listToString(s):
    
    # initialize an empty string
    str1 = "\n" 
    
    # return string  
    return (str1.join(s))
```

```python
# Returns df with the added column of details for plots

def column_value_in_next_line (df, columns):
    df['fragrance_display'] = ""
    
    for column in columns:
        df['fragrance_display'] += df[column] + '\n'
        
    return df
```

# Longevity

```python
df['longevity_in_hours'] = df['longevity'] / 60
```

```python
slope, intercept, r_value, p_value, std_err = sp.stats.linregress(df['number_of_sprays'].dropna(), df['longevity_in_hours'].dropna())
slope, intercept, r_value**2, p_value
```

```python
def get_reg_cof(df, columns, fragrance):
    temp_df = df[df['fragrance'] == fragrance].dropna(subset=[columns[0], columns[1]], axis=0)
    slope, intercept, r_value, p_value, std_err = sp.stats.linregress(temp_df[columns[0]], temp_df[columns[1]])
    return slope, intercept, r_value**2, p_value
```

```python
# get_reg_cof(df, ['number_of_sprays', 'longevity'], "Noir")
```

```python
def get_longevity(df, fragrance, sprays=5):
    slope, intercept, r_value_2, p_value = get_reg_cof(df, ['number_of_sprays', 'longevity'], fragrance)
    if np.isnan(slope):
        return -1
    else:
        return ( slope * sprays ) + intercept
```

```python
def get_projection(df, fragrance, sprays=5):
    slope, intercept, r_value_2, p_value = get_reg_cof(df, ['number_of_sprays', 'projection'], fragrance)
    if np.isnan(slope):
        return -1
    else:
        return ( slope * sprays ) + intercept
```

```python
def get_avg(df, column, fragrance):
    return df[df['fragrance'] == fragrance][column].mean()
```
